```python
#Selección de columnas de shape
#Seleccionadas (z-score)
selected_shape_z = ['eccentricity', 'extent', 'solidity']
df_shape_z_selected = df_shape_z[selected_shape_z]

#Seleccionadas (min-max)
selected_shape_mm = ['eccentricity', 'extent', 'solidity']
df_shape_mm_selected = df_shape_mm[selected_shape_mm]


#Selección de columnas de textura 
#Seleccionadas (z-score)
selected_textura_z = ['contrast', 'homogeneity', 'energy' ]
df_textura_z_selected = df_textura_z[selected_textura_z]

#Seleccionadas (min-max)
selected_textura_mm = ['contrast', 'homogeneity', 'energy' ]
df_textura_mm_selected = df_textura_mm[selected_textura_mm]


#Selección de columnas de fourier
#Seleccionadas (z-score)
selected_fourier_z = ['mean_magnitude', 'std_magnitude']
df_fourier_z_selected = df_fourier_z[selected_fourier_z]

#Seleccionadas (min-max)
selected_fourier_mm = ['mean_magnitude', 'std_magnitude']
df_fourier_mm_selected = df_fourier_mm[selected_fourier_mm]


#Selección de columnas de color
#Seleccionadas (z-score)
selected_color_z = ['brightness', 'hue', 'saturation', 'label', 'filename']
df_color_z_selected = df_color_z[selected_color_z]

#Seleccionadas (min-max)
selected_color_mm = ['brightness', 'hue', 'saturation', 'label', 'filename']
df_color_mm_selected = df_color_mm[selected_color_mm]



#combinación z-score
df_combined_all_z = pd.concat([df_shape_z_selected, df_textura_z_selected, df_fourier_z_selected, df_color_z_selected ], axis=1)

#combinación minmax
df_combined_all_mm = pd.concat([df_shape_mm_selected, df_textura_mm_selected, df_fourier_mm_selected, df_color_mm_selected ], axis=1)

# Guardar en archivos CSV
df_combined_all_z.to_csv(os.path.join(folder_path, "features_combined_all_zscore.csv"), index=False)
df_combined_all_mm.to_csv(os.path.join(folder_path, "features_combined_all_minmax.csv"), index=False)
```

```python
# Método del codo
WCSS = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=1000, random_state=42)
    kmeans.fit(df_combined_all_z_cl)
    WCSS.append(kmeans.inertia_)

# Graficar el codo
plt.plot(range(1, 11), WCSS, marker='o')
plt.xlabel('Número de Clusters (k)')
plt.ylabel('WCSS')
plt.title('Método del Codo para GLCM (z-score)')
plt.grid(True)
plt.show()
```

```python
# Creamos la figura con 2x2 subplots
fig, ax = plt.subplots(2, 2, figsize=(15, 10))

# Probamos con k = 2, 3, 4, 5
for i, k in enumerate([2, 3, 4, 5]):
    km = KMeans(n_clusters=k, init='k-means++', n_init=10, max_iter=300, random_state=42)
    
    # Posición del subplot
    row = i // 2
    col = i % 2
    
    # Creamos el visualizador y ajustamos los datos
    visualizer = SilhouetteVisualizer(km, colors='yellowbrick', ax=ax[row][col])
    visualizer.fit(df_combined_all_z_cl)  

plt.tight_layout()
plt.show()
```

```python
#Modelo k-means
cluster_all_z= KMeans(init='k-means++',n_clusters=3,random_state=42)
cluster_all_z.fit(df_combined_all_z_cl)   
```

```python
#Se añaden las etiquetas y vemos en una tabla de contigencia cuantos smaller o higher hay en cada cluster
df_combined_all_z['Cluster'] = cluster_all_z.labels_
pd.crosstab(df_combined_all_z['Cluster'], df_combined_all_z['label'])
```

Minmax (sin reducir la dimensionalidad con umap):

voy a probar a hace reduccion de la dimensionalidad con umap

```python
# 1. Reducción de dimensionalidad con UMAP
umap_model = umap.UMAP(n_neighbors=11, n_components=2, min_dist=0.01, random_state=42)
X_umap = umap_model.fit_transform(df_combined_all_z_cl)

# 2. Aplicar KMeans sobre el espacio reducido
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
clusters = kmeans.fit_predict(X_umap)

# 3. Añadir los clústeres al dataframe original
df_combined_all_z['Cluster_UMAP'] = clusters

# 4. Visualizar los resultados
plt.figure(figsize=(8, 6))
plt.scatter(
    X_umap[:, 0], X_umap[:, 1],
    c=clusters, cmap='Set1', edgecolor='k', alpha=0.7
)
plt.title('KMeans aplicado sobre datos reducidos con UMAP')
plt.xlabel('UMAP 1')
plt.ylabel('UMAP 2')
plt.colorbar(label='Clúster')
plt.grid(True)
plt.show()

# 5. Índice de silueta
silhouette = silhouette_score(X_umap, clusters)
print(f"Índice de silueta (UMAP + KMeans): {silhouette:.3f}")
```

```python
contingencia = pd.crosstab(df_combined_all_z['label'], df_combined_all_z['Cluster_UMAP'])
print(contingencia)
```

```python
# 1. Reducción de dimensionalidad con UMAP
umap_model_mm = umap.UMAP(n_neighbors=15, n_components=2, min_dist=0.01, random_state=42)
X_umap_mm = umap_model_mm.fit_transform(df_combined_all_mm_cl)

# 2. Aplicar KMeans sobre el espacio reducido
kmeans_mm = KMeans(n_clusters=3, init='k-means++', random_state=42)
clusters_mm = kmeans_mm.fit_predict(X_umap_mm)

# 3. Añadir los clústeres al dataframe original
df_combined_all_mm['Cluster_UMAP'] = clusters_mm

df_plot_mm = pd.DataFrame({
    'UMAP 1': X_umap_mm[:, 0],
    'UMAP 2': X_umap_mm[:, 1],
    'Cluster': clusters_mm.astype(str)
})

# Visualización con Seaborn
plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=df_plot_mm,
    x='UMAP 1',
    y='UMAP 2',
    hue='Cluster',
    palette='Set2',
    s=60,
    edgecolor='white',
    alpha=1,
    legend='full'
)

plt.title("Clusters KMeans (UMAP) All-vectors ")
plt.xlabel("UMAP componente 1")
plt.ylabel("UMAP componente 2")
plt.grid(True)
plt.legend(title='Clúster')
plt.show()

# 5. Índice de silueta
silhouette_mm = silhouette_score(X_umap_mm, clusters_mm)
print(f"Índice de silueta (UMAP + KMeans): {silhouette_mm:.3f}")
```

```python
contingencia = pd.crosstab(df_combined_all_mm['label'], df_combined_all_mm['Cluster_UMAP'])
print(contingencia)
```
